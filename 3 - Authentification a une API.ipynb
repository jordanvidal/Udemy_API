### Autres point d'acces (entrypoints)

```python
# huandu (access private user)
response = requests.get("http://api.github.com/users/huandu",
                       headers=headers)
huandu = response.json()
print(huandu)
```

```python
#Access organisation (FB)
response = requests.get("http://api.github.com/orgs/facebook",
                       headers=headers)
orgs = response.json()
print(orgs)
```

```python
#Access repository
response = requests.get("http://api.github.com/repos/octocat/Hello-World",
                       headers=headers)
training = response.json()
print(training)
```

## Pagination

```python
params = {"per_page":50, "page":1}
response = requests.get("http://api.github.com/users/rakeshsukla53/starred",
                        headers = headers,
                        params = params)
page1_repos = response.json()
print(page1_repos)
```

## Point d'acces User-Level

```python
response = requests.get("http://api.github.com/user",
                        headers=headers)
test = response.json()
print(test)
```
